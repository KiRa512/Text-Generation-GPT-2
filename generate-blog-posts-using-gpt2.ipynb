```python
from transformers import GPT2LMHeadModel , GPT2Tokenizer
```

```python
tokenizer = GPT2Tokenizer.from_pretrained('gpt2-large')
model = GPT2LMHeadModel.from_pretrained('gpt2-large',pad_token_id = tokenizer.eos_token_id)
```

```python
sentence = "I am Kakashi the copy ninja i know a thousand jutsu and now i will use them all and live up to my name"
input_ids = tokenizer.encode(sentence , return_tensors = "pt") # Giving each word a special identifier
```

```python
input_ids
```

```python
# you can go back to the original text using tokenizer.decoode
tokenizer.decode(input_ids[0][6])
```

```python
output = model.generate(input_ids, max_length = 150 , 
                        num_beams = 5,no_repeat_ngram_size = 2,
                        early_stoppings = True,)
```

```python
output
```

```python
tokenizer.decode(output[0],skip_special_tokens = True)
```

```python
#This is a good text actually, Now we'll import it to a text File
text = tokenizer.decode(output[0],skip_special_tokens = True)
```

```python
with open('NinjaWay.txt','w') as t:
    t.write(text)
```

# We'll now generate a longer blog

```python
output_long = model.generate(input_ids, max_length = 750 , 
                        num_beams = 5,no_repeat_ngram_size = 2,
                        early_stoppings = True,)
```

```python
tokenizer.decode(output_long[0],skip_special_tokens = True)
```

```python
l_text = tokenizer.decode(output_long[0],skip_special_tokens = True)
```

```python
with open('LongNinjaWay.txt','w') as t:
    t.write(l_text)
```
